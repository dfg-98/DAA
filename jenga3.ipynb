```python
def penga(h, C, E, M):
    if not isinstance(h, tuple):
        h = tuple(h)
    n = len(h)
    h_max = max(h)
    h_min = min(h)
    
    costs = dict()
    visited = {h}
    
    for i in range(h_max + 1):
        index = tuple(i for _ in range(n))
        costs[index] = 0
    # return costs
    return _penga(h, C, E, M, costs, visited, h_max, h_min, 0), costs


def _penga(h, C, E, M, costs, visited: set, h_max, h_min, count):
    n = len(h)
    
    if h in costs:
        return costs[h]
    else:
         print(f"{h=} {h_max=} {h_min=}")
    
    cost = np.inf
    min_index = 0
    max_index = 0
    l_min = h[0]
    l_max = h[0]
    for i in range(n):
        if h[i] < l_min:
            l_min = h[i]
            min_index = i
        else:
            l_max = h[i]
            max_index = i

    
    if l_min < h_max:
        i = min_index
        c_state = h[:i] + (h[i] + 1,) + h[i+1:]
        if not c_state in visited or c_state in costs:
            visited.add(c_state)
            cost = min(_penga(c_state, C, E, M, costs, visited, h_max, h_min, count+1) + C, cost)

    if l_max > h_min:
        i = max_index    
        e_state = h[:i] + (h[i] - 1,) + h[i+1:]
        if not e_state in visited or e_state in costs:
                    visited.add(e_state)
                    cost = min(_penga(e_state, C, E, M, costs, visited, h_max, h_min, count+1) + E, cost)
        
    if l_max > l_min + 1:
        i,j = max_index, min_index
        if i > j:
            m_state = h[:j] + (h[j] + 1,) + h[j+1:i] + (h[i] - 1,) + h[i+1:]
        else:
            m_state = h[:i] + (h[i] - 1,) + h[i+1:j] + (h[j] + 1,) + h[j+1:]
        if not m_state in visited or m_state in costs:
            visited.add(m_state)
            cost = min(_penga(m_state, C, E, M, costs, visited, h_max, h_min, count+1) + M, cost)
    
    costs[h] = cost
    return cost
```
